```python
import pandas as pd
import matplotlib.pyplot as plt
import random 
```

```python
df = pd.read_excel('C:\\Users\\mahat\\OneDrive\\Documents\\WorldPop.xlsx')
```

```python
print(df)
```

```python
df.set_index('Country Name', inplace=True)
```

```python
dates = df.columns
```

```python
print(dates)
```

```python
random_countries = random.sample(list(df.index), 10)
```

```python
for date in dates:
    plt.figure(figsize=(12, 6))
    df.loc[random_countries, date].plot(kind='bar')
    plt.title(f'Population distribution on {date}')
    plt.xlabel('Country')
    plt.ylabel('Population')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.savefig(f'population_distribution_{date}.png') 
    plt.show()
```
